Q.1 What is a random variable in probability theory?

Q.5  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

ANS: The difference between CDF and PDF lies in how they represent probabilities:

**PDF (Probability Distribution Function):**

* Used for continuous variables.

* Shows the probability density at each value.

* It tells you how likely a value is, but not the total probability up to that point.

* The area under the curve over an interval gives the probability.


  **CDF (Cumulative Distribution Function):**

* Used for both discrete and continuous variables.

* Shows the cumulative probability up to a certain value.

* It tells you the total probability of the variable being less than or equal to a specific value.

* The value of CDF increases from 0 to 1.

Q.2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for Discrete Uniform Distribution
low = 1    # Minimum value (inclusive)
high = 7   # Maximum value (exclusive)

# Generate discrete uniform random variables
sample_size = 10000
data = randint.rvs(low, high, size=sample_size)

# Unique values and their PMF (since uniform, all have equal probability)
x = np.arange(low, high)
pmf = [1 / (high - low)] * len(x)

# Plot PMF
plt.bar(x, pmf, width=0.6, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Probability')
plt.title('PMF of Discrete Uniform Distribution')
plt.xticks(x)
plt.ylim(0, max(pmf) + 0.05)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Q.3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome (0 or 1)
    p (float): Probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)


print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

uniform distribution>

Q.9 Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores of a dataset and plot them.

    Parameters:
    data (array-like): Input dataset

    Returns:
    z_scores (numpy array): Calculated Z-scores
    """
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.axhline(y=0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-score')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores


# Generate random data
sample_data = np.random.normal(50, 10, 100)

# Calculate and plot Z-scores
z = calculate_and_plot_zscores(sample_data)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
population_size = 100000
sample_size = 50
num_samples = 1000

# Step 1: Create a non-normal population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=population_size)

# Visualize the original population
plt.figure(figsize=(8, 5))
plt.hist(population, bins=50, color='skyblue', edgecolor='black', density=True)
plt.title('Original Non-Normal Population (Exponential Distribution)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Step 2: Sampling and calculating sample means
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size)
    sample_means.append(np.mean(sample))

# Step 3: Plot distribution of sample means
plt.figure(figsize=(8, 5))
plt.hist(sample_means, bins=30, color='lightgreen', edgecolor='black', density=True)
plt.title(f'Distribution of Sample Means (Sample Size = {sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)

# Optional: Overlay a normal distribution curve for reference
from scipy.stats import norm

mean_of_means = np.mean(sample_means)
std_of_means = np.std(sample_means)

x = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x, norm.pdf(x, mean_of_means, std_of_means), 'r--', label='Normal Curve')

plt.legend()
plt.show()
```

Q.19 Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

population_mean = 100
population_std = 15
sample = np.array([102, 98, 101, 99, 97, 103, 100, 96, 105, 99])


sample_mean = np.mean(sample)
n = len(sample)


z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z_stat:.3f}")


alpha = 0.05

z_critical = norm.ppf(1 - alpha/2)
print(f"Critical Z-value = ¬±{z_critical:.3f}")


if abs(z_stat) > z_critical:
    print("Result: Reject the null hypothesis (H‚ÇÄ). Significant difference detected.")
else:
    print("Result: Fail to reject the null hypothesis (H‚ÇÄ). No significant difference detected.")
```

Q.20 Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats


sample_data = [50, 52, 53, 49, 48, 51, 54, 50, 52, 53]
n = len(sample_data)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# Confidence level
confidence = 0.95
alpha = 1 - confidence

# t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)

# Margin of error
margin_error = t_critical * (sample_std / np.sqrt(n))

# Confidence interval
lower_limit = sample_mean - margin_error
upper_limit = sample_mean + margin_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Sample Standard Deviation = {sample_std:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({lower_limit:.2f}, {upper_limit:.2f})")
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Generate data
np.random.seed(0)
mu = 50
sigma = 10
n = 100

data = np.random.normal(mu, sigma, n)  # Generate n data points

# Step 2: Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# Step 3: Calculate Confidence Interval
confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_error = t_critical * (sample_std / np.sqrt(n))

lower_limit = sample_mean - margin_error
upper_limit = sample_mean + margin_error

print(f"Sample Mean = {sample_mean:.2f}")
print(f"Sample Standard Deviation = {sample_std:.2f}")
print(f"{int(confidence*100)}% Confidence Interval = ({lower_limit:.2f}, {upper_limit:.2f})")

# Step 4: Plotting
plt.figure(figsize=(10,6))
plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Sample Mean = {sample_mean:.2f}")
plt.axvline(lower_limit, color='green', linestyle='--', label=f"Lower Limit = {lower_limit:.2f}")
plt.axvline(upper_limit, color='green', linestyle='--', label=f"Upper Limit = {upper_limit:.2f}")
plt.title('Histogram of Generated Normal Data with Confidence Interval')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

Q.22 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu = 0
sigma = 1

x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate PDF values
pdf = norm.pdf(x, mu, sigma)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label=f'Normal Distribution\nMean = {mu}, Std Dev = {sigma}')
plt.fill_between(x, pdf, color='skyblue', alpha=0.4)
plt.title('Probability Density Function (PDF) of Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

Q.23  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson


lambda_value = 3  # Average rate of events (mean)

# Range of k values (number of occurrences)
k_values = np.arange(0, 11)

cdf_values = poisson.cdf(k_values, mu=lambda_value)


for k, cdf in zip(k_values, cdf_values):
    print(f"P(X ‚â§ {k}) = {cdf:.4f}")


plt.figure(figsize=(10, 6))
plt.step(k_values, cdf_values, where='post', label=f'Poisson CDF (Œª={lambda_value})', color='blue')
plt.xlabel('Number of Events (k)')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Poisson Distribution')
plt.grid(True)
plt.legend()
plt.show()
```

Q.26 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate dataset from normal distribution
np.random.seed(0)
mean = 100
std_dev = 15
sample_size = 1000

data = np.random.normal(mean, std_dev, sample_size)

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")

# Step 4: Visualization
plt.figure(figsize=(10, 6))
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Boxplot of Normally Distributed Data')
plt.xlabel('Value')
plt.grid(True)
plt.show()
```

Q.27 Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(1)
data = np.random.normal(loc=50, scale=10, size=1000)  # Normal distribution (mean=50, std=10)

mean = np.mean(data)
std_dev = np.std(data)
z_scores = (data - mean) / std_dev

print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"Normalized Data Mean: {np.mean(z_scores):.2f}, Std Dev: {np.std(z_scores):.2f}")

# Visualization
plt.figure(figsize=(12, 5))

# Original Data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Normalized Data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data Distribution')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Q.28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std_dev=1, size=1000):
    # Generate dataset from normal distribution
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data, fisher=False)  # Set fisher=False for Pearson's kurtosis (normal ‚âà 3)

    print(f"Skewness of dataset: {data_skewness:.4f}")
    print(f"Kurtosis of dataset: {data_kurtosis:.4f}")

    return data_skewness, data_kurtosis


calculate_skewness_kurtosis(mean=50, std_dev=10, size=1000)
```
